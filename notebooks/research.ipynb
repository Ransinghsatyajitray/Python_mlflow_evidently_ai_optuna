```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))

numerical_columns=iris.select_dtypes(exclude="object").columns


for col in numerical_columns:
    sns.histplot(data=iris,x=col,kde=True)
    print("\n")
    plt.show()
```

```python

# Understanding the collinearity
sns.heatmap(iris[numerical_columns].corr(),annot=True)
```
